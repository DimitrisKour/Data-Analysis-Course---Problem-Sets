```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

# 'Ασκηση 5:

# 1 - Gaussian:

```python
# Παράμετροι: (μ, σ)
gauss_params = [(0,1), (0,2), (5,1)]
gauss_sizes = [100, 1000, 5000]
gauss_bins = 30

for mu, sigma in gauss_params:
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,6), sharex=True, sharey=True)
    fig.suptitle(f'Κανονική Κατανομή μ={mu}, σ={sigma}', fontsize=14)
    
    for ax, size in zip(axes, gauss_sizes):
        data = np.random.normal(loc=mu, scale=sigma, size=size)
        ax.hist(data, bins=gauss_bins, color='skyblue', edgecolor='black', alpha=0.7, label=f'N={size}')
        ax.set_ylabel("Συχνότητα")
        ax.legend()
    
    axes[-1].set_xlabel("Τιμή")
    plt.tight_layout(rect=[0,0,1,0.96])
    plt.show()
```

```python
# Κανονική κατανομή, παράδειγμα
mu, sigma = 0, 1
N = 1000
data = np.random.normal(mu, sigma, N)

# Διαφορετικά bins για υποplots
bins_list = [10, 30, 50]

fig, axes = plt.subplots(1, 3, figsize=(16,6), sharey=True)
fig.suptitle(f'Κανονική Κατανομή μ={mu}, σ={sigma}, N={N} - Επίδραση των bins', fontsize=16)

for ax, b in zip(axes, bins_list):
    ax.hist(data, bins=b, color='skyblue', edgecolor='black', alpha=0.7)
    ax.set_title(f'bins={b}')
    ax.set_xlabel("Τιμή")
    ax.set_ylabel("Συχνότητα")

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```

# 2 - Uniform:

```python
# Παράμετροι: (low, high)
uniform_params = [(0,1), (0,10), (-5,5)]
uniform_sizes = [100, 1000, 5000]
uniform_bins = 30

for low, high in uniform_params:
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,6), sharex=True, sharey=True)
    fig.suptitle(f'Ομοιόμορφη Κατανομή [{low},{high}]', fontsize=14)
    
    for ax, size in zip(axes, uniform_sizes):
        data = np.random.uniform(low=low, high=high, size=size)
        ax.hist(data, bins=uniform_bins, color='lightgreen', edgecolor='black', alpha=0.7, label=f'N={size}')
        ax.set_ylabel("Συχνότητα")
        ax.legend()
    
    axes[-1].set_xlabel("Τιμή")
    plt.tight_layout(rect=[0,0,1,0.96])
    plt.show()
```

```python
low, high = 0, 10
N = 1000
data = np.random.uniform(low, high, N)

bins_list = [5, 20, 40]

fig, axes = plt.subplots(1, 3, figsize=(16,6), sharey=True)
fig.suptitle(f'Ομοιόμορφη Κατανομή [{low},{high}], N={N} - Επίδραση των bins', fontsize=16)

for ax, b in zip(axes, bins_list):
    ax.hist(data, bins=b, color='lightgreen', edgecolor='black', alpha=0.7)
    ax.set_title(f'bins={b}')
    ax.set_xlabel("Τιμή")
    ax.set_ylabel("Συχνότητα")

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```

# 3 - Exponential:

```python
# Παράμετροι: scale = 1/λ
exp_params = [1, 2, 5]
exp_sizes = [100, 1000, 5000]
exp_bins = 30

for scale in exp_params:
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16,6), sharex=True, sharey=True)
    fig.suptitle(f'Εκθετική Κατανομή scale={scale}', fontsize=14)
    
    for ax, size in zip(axes, exp_sizes):
        data = np.random.exponential(scale=scale, size=size)
        ax.hist(data, bins=exp_bins, color='salmon', edgecolor='black', alpha=0.7, label=f'N={size}')
        ax.set_ylabel("Συχνότητα")
        ax.legend()
    
    axes[-1].set_xlabel("Τιμή")
    plt.tight_layout(rect=[0,0,1,0.96])
    plt.show()
```

```python
scale = 2  # mean=scale
N = 1000
data = np.random.exponential(scale, N)

bins_list = [10, 30, 50]

fig, axes = plt.subplots(1, 3, figsize=(16,6), sharey=True)
fig.suptitle(f'Εκθετική Κατανομή scale={scale}, N={N} - Επίδραση των bins', fontsize=16)

for ax, b in zip(axes, bins_list):
    ax.hist(data, bins=b, color='salmon', edgecolor='black', alpha=0.7)
    ax.set_title(f'bins={b}')
    ax.set_xlabel("Τιμή")
    ax.set_ylabel("Συχνότητα")

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```

# Gaussian + Uniform:

```python
# Κανονική + Ομοιόμορφη
gauss_data = np.random.normal(0,1,1000)
uniform_data = np.random.uniform(0,10,1000)
combined_data = np.concatenate([gauss_data, uniform_data])

fig, ax = plt.subplots(figsize=(8,5))
ax.hist(combined_data, bins=30, color='violet', edgecolor='black', alpha=0.7, label='Ένωση')
ax.set_title("Ένωση Κανονικής και Ομοιόμορφης")
ax.set_xlabel("Τιμή")
ax.set_ylabel("Συχνότητα")
ax.legend()
plt.show()
```

# Add random variables:

```python
# Κανονική + Εκθετική
sum_data = gauss_data + np.random.exponential(scale=2, size=1000)

fig, ax = plt.subplots(figsize=(8,5))
ax.hist(sum_data, bins=30, color='orange', edgecolor='black', alpha=0.7, label='Άθροισμα')
ax.set_title("Άθροισμα Κανονικής + Εκθετικής")
ax.set_xlabel("Τιμή")
ax.set_ylabel("Συχνότητα")
ax.legend()
plt.show()
```

# Συμπεράσματα από τα ιστογράμματα

1. **Παράμετροι κατανομής**:
   - Η μέση τιμή (μ) μετακινεί την κατανομή δεξιά ή αριστερά.
   - Η τυπική απόκλιση (σ) διαστέλλει ή συστέλλει την κατανομή.

2. **Μέγεθος δείγματος**:
   - Μικρό δείγμα → ακανόνιστο, θορυβώδες ιστογράφημα.
   - Μεγάλο δείγμα → πιο ομαλό και αντιπροσωπευτικό σχήμα.

3. **Bin width / number of bins**:
   - Λίγα bins → χονδροειδές ιστογράφημα.
   - Πολλά bins → θορυβώδες, υπερβολικά λεπτομερές ιστογράφημα.

4. **Ένωση δύο κατανομών**:
   - Το ιστογράφημα συνδυάζει τα χαρακτηριστικά και των δύο κατανομών.
   - Μπορεί να εμφανιστούν δύο κορυφές ή πιο επίπεδη περιοχή.

5. **Άθροισμα δύο τυχαίων μεταβλητών**:
   - Δημιουργεί νέα κατανομή με διαφορετικό σχήμα.
   - Αν μια κατανομή είναι συμμετρική και η άλλη ασύμμετρη, το αποτέλεσμα μπορεί να έχει skewed κατανομή.
   - Μεγαλύτερα δείγματα δίνουν πιο ομαλό αποτέλεσμα.


# Άσκηση 6:

```python
N = 250
steps_per_walk = 2 * N  # 500 βήματα
num_walks = 5

# Δημιουργία 5 τυχαίων περιπάτων
walks = []
for _ in range(num_walks):
    # Κάθε βήμα: +1 ή -1 με πιθανότητα 0.5
    steps = np.random.choice([-1, 1], size=steps_per_walk)
    # Συνεχής θέση
    walk = np.cumsum(steps)
    walks.append(walk)
```

```python
plt.figure(figsize=(12,6))

for i, walk in enumerate(walks, 1):
    plt.plot(range(steps_per_walk), walk, label=f'Walk {i}')

plt.title('5 Τυχαίοι Μονοδιάστατοι Περιπάτοι (Random Walks)')
plt.xlabel('Βήματα')
plt.ylabel('Θέση x')
plt.legend()
plt.grid(True)
plt.show()
```

```python
num_walks_large = 1000

final_positions = []

for _ in range(num_walks_large):
    steps = np.random.choice([-1,1], size=steps_per_walk)
    final_pos = np.sum(steps)  # ή np.cumsum(steps)[-1]
    final_positions.append(final_pos)

final_positions = np.array(final_positions)

# Ιστόγραμμα
plt.figure(figsize=(12,6))
plt.hist(final_positions, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title(f'Ιστόγραμμα Τελικών Θέσεων από {num_walks_large} Τυχαίους Περιπάτους')
plt.xlabel('Τελική θέση x')
plt.ylabel('Συχνότητα')
plt.grid(True)
plt.show()
```

# Παρατηρήσεις από τα γραφήματα των τυχαίων περιπάτων

1. **5 Τυχαίοι Περιπάτοι**:
   - Κάθε περίπατος ξεκινά από 0.
   - Υπάρχει μεγάλη διαφοροποίηση μεταξύ των περιπάτων λόγω τυχαίας επιλογής βημάτων.
   - Η γραμμή εμφανίζει **φτωχές κεντρικές τάσεις**, αλλά συνολικά κινείται γύρω από 0.

2. **1000 Περιπάτοι – Τελικές Θέσεις**:
   - Η κατανομή των τελικών θέσεων είναι περίπου **κανονική (Gaussian)**.
   - Η μέση τιμή περίπου 0, συμμετρική κατανομή.
   - Το εύρος των τελικών θέσεων αυξάνει με τον αριθμό βημάτων (~sqrt(N)).

3. **Γενικό σχόλιο**:
   - Ο μονοδιάστατος τυχαίος περίπατος ακολουθεί θεωρητικά **Κανονική κατανομή για μεγάλους αριθμούς βημάτων** λόγω του Κεντρικού Ορίου.
   - Τα γραφήματα οπτικοποιούν τη διάχυση γύρω από την αρχική θέση.
